```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\Priyanshu\Desktop\Titanic Train.csv")
```

```python
df.Age.fillna(df.Age.mean(),inplace=True)
df.Embarked.fillna('S')
df.describe()
```

```python
del df['Name'],df['Ticket'],df['Cabin']
```

```python
def f(s):
    if s=='male':
        return 1
    else:
        return 0

df.Sex=df.Sex.apply(f)
```

```python
del df['Parch'],df['Embarked'],
df.head()
```

```python
y_train=df.iloc[:,5]
```

```python
y_train.head()
```

```python
# X=np.r_[0:5,6:11]
x_train=df.iloc[:,0:5]
x_train.head()
```

```python
cf=pd.read_csv(r"C:\Users\Priyanshu\Desktop\Titanic test.csv")
cf.Age.fillna(cf.Age.mean(),inplace=True)
cf.Embarked.fillna('S')
del cf['Name'],cf['Ticket'],cf['Cabin'],cf['Parch'],cf['Embarked']

```

```python
cf.Sex=cf.Sex.apply(f)
x_test=cf.iloc[:]
cf.head()
x_test.head()
```

```python
from sklearn.preprocessing import StandardScaler
clf=StandardScaler()
x_train=clf.fit_transform(x_train)
x_test=clf.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
lr=LogisticRegression()
grid={'C':[0.08,0.09,0.1,0.2,0.3]}
abc=GridSearchCV(lr,grid,cv=17)
abc.fit(x_train,y_train)
abc.best_estimator_
```

```python
lr=LogisticRegression(C=0.1)
lr.fit(x_train,y_train)
```

```python
ans=lr.predict(x_test)
```
